```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Perseus Dwarfs

Using the same process for merging datasets as we did with NGC 383 and 507.

```python
perseus_sdss = pd.read_csv('../datasets/sdss/perseus.csv')
perseus_sdss = perseus_sdss[perseus_sdss.type!='STAR']
perseus_sdss.head()
```

```python
# No photometric redshifts in file
perseus_ned = pd.read_csv('../datasets/ned/perseus.csv')
perseus_ned.head()
```

```python
# Final table of ngc507, cross-id table with redshifts from NED
perseus = pd.merge(perseus_sdss, perseus_ned[['ObjectName', 'Redshift', 'RedshiftFlag']], on='ObjectName')
perseus.head() # 183 galaxies in total
```

```python
# Export merged dataset
perseus.to_csv(r'../datasets/complete/perseus.csv', index=False)
```

### Perseus Background Giants

```python
perseus_back_sdss = pd.read_csv('../datasets/sdss/perseus-back.csv')
perseus_back_sdss = perseus_back_sdss[perseus_back_sdss.type!='STAR']
perseus_back_sdss.head()
```

```python
perseus_back_ned = pd.read_csv('../datasets/ned/background.csv')
perseus_back_ned = perseus_back_ned[perseus_back_ned.RedshiftFlag!='PHOT']
perseus_back_ned.head()
```

```python
perseus_back = pd.merge(perseus_back_sdss, perseus_back_ned[['ObjectName', 'Redshift', 'RedshiftFlag']], on='ObjectName')
perseus_back.head() # 55 galaxies with spectroscopic redhifts
```

```python
# Export merged list
perseus_back.to_csv('../datasets/complete/perseus-back.csv', index=False)
```
