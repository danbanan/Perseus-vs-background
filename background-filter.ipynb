```python
import pandas as pd
```

```python
ngc383_all = pd.read_csv('./complete/ngc383.csv')
ngc383_all.head()
```

```python
# Filter out galaxies in the foreground of NGC 383
ngc383_back = ngc383_all.loc[(ngc383_all['Redshift'] < 0.0125) | (ngc383_all['Redshift'] > 0.0225)]
ngc383_back.head()
```

```python
ngc383_back = ngc383_back.loc[(ngc383_back['modelMag_r'] > 15.5) & (ngc383_back['modelMag_r'] < 16.5)]
ngc383_back.head()
# 27 background giants of NGC 383
```

```python
ngc507_all = pd.read_csv('./complete/ngc507.csv')
ngc507_all.head()
```

```python
# Filter out galaxies in the foreground of NGC 507
ngc507_back = ngc507_all.loc[(ngc507_all['Redshift'] < 0.0125) | (ngc507_all['Redshift'] > 0.0225)]
ngc507_back.head()
```

```python
ngc507_back = ngc507_back.loc[(ngc507_back['modelMag_r'] > 15.5) & (ngc507_back['modelMag_r'] < 16.5)]
ngc507_back.head()
# 4 background giants of NGC 507
```

```python
background = ngc383_back.append(ngc507_back, ignore_index=True)
background.head()
# 31 background galaxies in total
```

```python
# Export appended list
background.to_csv(r'./complete/background.csv', index=False)
```
