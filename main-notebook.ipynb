```python
ngc383_crossid = pd.read_csv('./training/ngc383-crossid.csv')
ngc383_crossid = ngc383[ngc383.type!='STAR']
ngc383_crossid.head()
```

```python
ngc383_ned = pd.read_csv('./training/ngc383-arc60_nospace.csv')
ngc383_ned = ngc383_ned[ngc383_ned.RedshiftFlag!='PHOT']
ngc383_ned.head()
```

```python
# Final table of ngc383, cross-id table with redshifts from NED
ngc383 = pd.merge(ngc383_crossid, ngc383_ned[['ObjectName', 'Redshift', 'RedshiftFlag']], on='ObjectName')
ngc383.head()
```
