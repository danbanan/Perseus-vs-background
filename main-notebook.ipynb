```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ngc383_crossid = pd.read_csv('./training/ngc383-crossid.csv')
ngc383_crossid = ngc383[ngc383.type!='STAR']
ngc383_crossid.head()
```

```python
ngc383_ned = pd.read_csv('./training/ngc383-arc60_nospace.csv')
ngc383_ned = ngc383_ned[ngc383_ned.RedshiftFlag!='PHOT']
ngc383_ned.head()
```

```python
# Final table of ngc383, cross-id table with redshifts from NED
ngc383 = pd.merge(ngc383_crossid, ngc383_ned[['ObjectName', 'Redshift', 'RedshiftFlag']], on='ObjectName')
ngc383.head()
```

```python
ngc507_crossid = pd.read_csv('./training/ngc507-crossid.csv')
ngc507_crossid = ngc507[ngc507.type!='STAR']
ngc507_crossid.head(5)
```

```python
ngc507_ned = pd.read_csv('./training/ngc507-arc60_nospace.csv')
ngc507_ned = ngc507_ned[ngc507_ned.RedshiftFlag!='PHOT']
ngc507_ned.head()
```

```python
# Final table of ngc507, cross-id table with redshifts from NED
ngc507 = pd.merge(ngc507_crossid, ngc507_ned[['ObjectName', 'Redshift', 'RedshiftFlag']], on='ObjectName')
ngc507.head()
```

```python
perseus_crossid = pd.read_csv('./training/perseus_CROSSID_.05Radius_RangeR.csv')
perseus_crossid.head()
```

```python
# No photometric redshifts in file
perseus_ned = pd.read_csv('./training/NED_perseus.csv')
perseus_ned.head()
```

```python
hist = pd.DataFrame({'Perseus': perseus.modelMag_r, 'NGC 383': ngc383.modelMag_r, 'NGC 507': ngc507.modelMag_r}, columns=['Perseus','NGC 383','NGC 507'])
hist.plot.hist(alpha=0.5, stacked='True', figsize=(10,10))
```
