### Data Analysis of Magnitudes, Redshifts and Color
A datasets of 56 dwarf galaxies is far from enough to train an accurate classification model. We want to increase the dataset by performing an all-sky query in SDSS. We will filter the search by similar spectroscopic redshifts, magnitudes and color. Before we can do so, we need to do some data analysis to find the distribution of the paramaters mentioned. We already know that the range of the magnitude should be 15.5-16.5 . We need to find the range for redshifts and color.

````python
import matplotlib.pyplot as plt
````

````python
color = dwarfs['u0']-dwarfs['z0']
color.head()
````

````python
fig, axes = plt.subplots(ncols=3, figsize=(15, 8))
dwarfs['r0'].plot.hist(ax=axes[0], bins=15)
dwarfs['Redshift'].plot.hist(ax=axes[1], bins=15)
color.plot.hist(ax=axes[2], bins=15)
axes[0].set_title('Magnitude', fontsize=12)
axes[1].set_title('Redshift', fontsize=12)
axes[2].set_title('$u-z\:[\mathrm{Mag}]$', fontsize=12)
plt.show()
````

We can't use the redshift range from Perseus, NGC 383 or 507 redshift range due to the velocity within the cluster. Instead we will be looking at redshift of the center of each cluster and use median +/- 10% .

Center redhifts of:
* NGC 507 $\rightarrow$ 0.01646 +/- 0.00002
* NGC 383 $\rightarrow$ 0.01700 +/- 0.00004
* Perseus $\rightarrow$ 0.01790 

$0.017\times0.9 = 0.0153$

$0.017\times1.1 = 0.0187$ 

Redshift range: $[0.0153-0.0187]$

````python
color.max()
````

Looking at the third histogram, the color range should be between 2.5 and 3.4 .

**All-sky search for dwarfs:**
```SQL
SELECT
    p.objid, p.ra,p.dec, p.run, p.rerun, p.camcol, p.field, 
    s.class AS type, (p.u-p.extinction_u) AS u0, 
    (p.g-p.extinction_g) AS g0, (p.r-p.extinction_r) AS r0, 
    (p.i-p.extinction_i) AS i0, (p.z-p.extinction_z) AS z0, 
    s.z AS Redshift
FROM PhotoObj AS p
    JOIN SpecObj AS s ON s.bestobjid = p.objid
WHERE 
    p.modelMag_r BETWEEN 15 AND 17.5
    AND s.z BETWEEN 0.0153 AND 0.0187
    AND ((p.u-p.extinction_u)-(p.z-p.extinction_z)) BETWEEN 2.5 AND 3.4
```
