```python
import pandas as pd
```

```python
ngc383_all = pd.read_csv('./complete/ngc383.csv')
ngc383_all.head()
```

```python
#df.loc[(df['column_name'] >= A) & (df['column_name'] <= B)]
ngc383 = ngc383_all.loc[(ngc383_all['Redshift'] > 0.0125) & (ngc383_all['Redshift'] < 0.0225)]
ngc383.head()
# 68 foreground galaxies in total
```

```python
ngc383_dwarf = ngc383.loc[(ngc383['modelMag_r'] > 15.5) & (ngc383['modelMag_r'] < 16.5)]
ngc383_dwarf.head()
# 18 foreground dwarfs in total
```

```python
ngc507_all = pd.read_csv('./complete/ngc507.csv')
ngc507_all.head()
```

```python
ngc507 = ngc507_all.loc[(ngc507_all['Redshift'] > 0.0125) & (ngc507_all['Redshift'] < 0.0225)]
ngc507.head()
# 65 foreground galaxies in total
```

```python
ngc507_dwarf = ngc507.loc[(ngc507['modelMag_r'] > 15.5) & (ngc507['modelMag_r'] < 16.5)]
ngc507_dwarf.head()
# 8 foreground dwarfs in total
```

```python
dwarfs = ngc383_dwarf.append(ngc507_dwarf, ignore_index=True)
dwarfs.head()
# 26 dwarfs in total
```

```python
# Export appended list
dwarfs.to_csv(r'./complete/dwarfs.csv', index=False)
```
