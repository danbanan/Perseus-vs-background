```python
import pandas as pd
```

### NGC 383

```python
ngc383 = pd.read_csv('./complete/ngc383.csv')
ngc383.head()
```

```python
#df.loc[(df['column_name'] >= A) & (df['column_name'] <= B)]
# Filter the foreground galaxies of NGC 383
ngc383 = ngc383.loc[(ngc383['Redshift'] > 0.0125) & (ngc383['Redshift'] < 0.0225)]
ngc383.head() # 68 foreground galaxies in total
```

```python
ngc383_dwarf = ngc383.loc[(ngc383['modelMag_r'] > 15.5) & (ngc383['modelMag_r'] < 16.5)]
ngc383_dwarf.head() # 18 foreground dwarfs in total
```

### NGC 507

```python
ngc507 = pd.read_csv('./complete/ngc507.csv')
ngc507.head()
```

```python
ngc507 = ngc507.loc[(ngc507['Redshift'] > 0.0125) & (ngc507['Redshift'] < 0.0225)]
ngc507.head() # 65 foreground galaxies in total
```

```python
ngc507_dwarf = ngc507.loc[(ngc507['modelMag_r'] > 15.5) & (ngc507['modelMag_r'] < 16.5)]
ngc507_dwarf.head() # 8 foreground dwarfs in total
```

### Perseus

```python
perseus = pd.read_csv('./complete/perseus.csv')
perseus.head() # This dataset is already filtered by redshift between 0.01 and 0.03 in NED search
```

```python
perseus_dwarfs = perseus.loc[(perseus['modelMag_r'] > 15.5) & (perseus['modelMag_r'] < 16.5)]
perseus_dwarfs.head() # 30 dwarfs in total
```

```python
dwarfs = ngc383_dwarf.append(ngc507_dwarf, ignore_index=True)
dwarfs = dwarfs.append(perseus_dwarfs, ignore_index=True)
str = 'Number of dwarf galaxies: ' + str(len(dwarfs.index))
dwarfs.head()
```

### Data Analysis of Magnitudes, Redshifts and Color
A datasets of 56 dwarf galaxies is far from enough to train an accurate classification model. We want to increase the dataset by performing an all-sky query in SDSS. We will filter the search by similar spectroscopic redshifts, magnitudes and color. Before we can do so, we need to do some data analysis to find the distribution of the paramaters mentioned. We already know that the range of the magnitude should be 15.5-16.5, and spectroscopic redshifts 0.01-0.03. Redshift range from NGC 383, 507 and Perseus fall within this range.

```python
import matplotlib.pyplot as plt
```

```python
color = dwarfs['modelMag_u']-dwarfs['modelMag_z']
color.head()
```

```python
fig, axes = plt.subplots(ncols=3, figsize=(15, 8))
dwarfs['modelMag_r'].plot.hist(ax=axes[0], bins=15)
dwarfs['Redshift'].plot.hist(ax=axes[1], bins=15)
color.plot.hist(ax=axes[2], bins=15)
axes[0].set_title('Magnitude', fontsize=12)
axes[1].set_title('Redshift', fontsize=12)
axes[2].set_title('$u-z\:[\mathrm{Mag}]$', fontsize=12)
plt.show()
```

Center redhifts of:
* NGC 507 $\rightarrow$ 0.01646 +/- 0.00002
* NGC 383 $\rightarrow$ 0.01700 +/- 0.00004
* Perseus $\rightarrow$ 0.01790 

The redshift range we will use is the median +/- 10% .

$0.017\times0.9 = 0.0153$

$0.017\times1.1 = 0.0187$ 

Redshift range: $[0.0153-0.0187]$
