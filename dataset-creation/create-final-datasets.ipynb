From all the PhotoObj parameters, for the first run we'll be using all parameters that are not RA, DEC and Errors.
Further we have edited the PhotoObj parameter to only include the parameters that are usefull for our classification attempt with SVM.

````python
#Opening the selected param text file 
svmParam = open('../datasets/params/sdss-params-nl.txt','r')
    
svmList = svmParam.readlines()
svmParam.close()

#Convert the selected param text file into a more readible format for
#crossID which wants no new lines containing only a ',' to sepereate
#parameters
svmParam_no_nl = open("../datasets/params/sdss-params-no-nl.txt","w+")

for n in svmList:
    svmParam_no_nl.write(n.replace('\n',''))

svmParam_no_nl.readlines()
    
svmParam_no_nl.close()
````
